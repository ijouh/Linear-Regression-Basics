```python
def fonction_cout(x,y,w,b):
    m=len(x)
    j=0
    for i in range(m):
        j=j+(w*x[i]+b-y[i])**2
    return (1/2*m)*j
```

```python
fonction_cout(x,y,w,b)
```

```python
def deriver(x, y, w, b):
    m = len(x)
    j_w = 0
    j_b = 0
    for i in range(m):
        j_w += (w * x[i] + b - y[i]) * x[i]
        j_b += (w * x[i] + b - y[i])
    return j_w / m, j_b / m

def gradient_descente(x, y, w, b, alpha):
    max_iterations = 10000  # Limite le nombre d'itérations pour éviter une boucle infinie
    epsilon = 0.00001  # Seuil de convergence
    iteration = 0

    while iteration < max_iterations:
        j_w, j_b = deriver(x, y, w, b)
        w_temp = w - alpha * j_w
        b_temp = b - alpha * j_b

        if abs(w - w_temp) < epsilon and abs(b - b_temp) < epsilon:
            break

        w = w_temp
        b = b_temp
        iteration += 1

    return w, b
```

```python
w,b=gradient_descente(x,y,w,b,0.01)
```
